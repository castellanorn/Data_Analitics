<font color=royalblue>1. Calculadora del √≠ndice de masa corporal</font>  

Escribe una funci√≥n que calcule el IMC ingresado por el usuario/a, es decir, quien lo ejecute deber√° ingresar estos datos. Puedes obtener m√°s informaci√≥n de su c√°lculo en: üîó √çndice de masa corporal IMC que es y c√≥mo se calcula.  
La funci√≥n debe clasificar el resultado en sus respectivas categor√≠as.  
Consejo: Intenta validar los datos previamente, para que env√≠e un mensaje de advertencia si los datos introducidos por el usuario/a no est√°n en el formato adecuado o no toma valores razonables.

<font color="royalblue">Objetivo - IMC:</font>  
- F√≥rmula para calculo de *IMC* = *peso* en kg / *altura*^2 en metros,
- Solicitar valores para el c√°lculo con **print**,
- Validar datos de entrada con **float** para que acepte num√©ricos y con **abs** para convertir a positivos, 
- Enviar mensaje con **try-except** para indicar error en el ingreso de datos, esto cuando se pide valor de *peso* y despu√©s cuando se pide el valor *altura*, para el √∫ltimo se indica el separador de decimal como punto,
- Esperar nuevamente los datos en caso de enviar mensaje de ERROR,
- Redondear de resultado *IMC* a 2 decimales con **round**,
- Establecer los rango de *IMC* con condicional **if-elif-else**, seg√∫n valor de IMC. 
    -   Bajo peso: menos de 18,5. 
    -   Peso normal: 18,5-24,9. 
    -   Sobrepeso: 25-29,9. 
    -   Obesidad: m√°s de 30.

```python
def calculoIMC():
    print("Para definir tu Indice de Masa Corporal (IMC), necesitamos algunos datos.")      
    try:                                                                                    
        peso =abs(float(input("Cual es tu peso actual? (unidades en kg)")))                 # Solicita peso
    except ValueError:                                                                      # Manejo de error si no es numerico      
        print("ERROR, ingresa un valor num√©rico para el peso.")
        peso = abs(float(input()))                                                          # Espera entrada nuevamente
    try:
        altura = abs(float(input("Cual es tu altura actual? (unidades en metros)")))    # Solicita altura
    except ValueError:                                                                  # Manejo de error si no es numerico, con punto decimal                                           
        print("ERROR, ingresa un valor num√©rico para la altura. Separador decimal con punto.")  
        altura = abs(float(input()))                                                    # Espera entrada nuevamente     
    imc = round(peso / altura**2)                                                           # Calculo del IMC       
    print("Tu IMC actual es de: "+ str(imc) +". Seg√∫n este valor:")
    if imc < 18.5:                                                                          # Condicionales para rangos del IMC                              
        print ("Tienes peso bajo")
    elif 18.5 <= imc < 24.9:
        print ("Tienes un peso normal")
    elif 25 <= imc < 29.9:
        print ("Tienes sobrepeso")
    else:
        print ("Tienes obesidad, debes bajar de peso")

calculoIMC()
```

<font color=royalblue>2. Conversor de temperaturas</font>  

Existen diversas unidades de temperatura utilizadas en distintos contextos y regiones. Las m√°s comunes son Celsius (¬∞C), Fahrenheit (¬∞F) y Kelvin (K). Tambi√©n existen otras unidades como Rankine (¬∞Ra) y R√©aumur (¬∞Re).  
Selecciona al menos 2 conversores, de modo que al introducir una temperatura devuelva, como m√≠nimo, dos conversiones, de modo que se puedan guardar (recuerda que un print() no se puede guardar nunca).
Consejo: Intenta validar los datos previamente, para que env√≠e un mensaje de advertencia si los datos introducidos por el usuario/a no est√°n en el formato adecuado.  
(EXTRA): Piensa una manera de almacenar todas las posibles conversiones en un solo objeto (¬øLista? ¬øDiccionario? ¬øDataFrame?) en vez de escribir muchos if else en funci√≥n de la temperatura de origen y la temperatura de destino.

<font color=royalblue>Objetivo - CONVERTIR TEMP:</font>
- F√≥rmulas de para realizar la conversi√≥n de temperatura:
    -   Celsius a Fahrenheit	T ( ¬∞ F ) = [ 9 5 T ( ¬∞ C ) ] + 32 ,
    -   Farenheit a Celsius 	T ( ¬∞ C ) = 5 9 [ T ( ¬∞ F ) ‚àí 32 ], ‚Äç
    -   Celsius a Kelvin	    T ( K ) = T ( ¬∞ C ) + 273,
    -   Kelvin a Celsius	    T ( ¬∞ C ) = T ( K ) ‚àí 273, ‚Äç
- Introducir objetivo de la funcion con un **print**,
- Validar dato *temp* con **float** para que acepte valor num√©rico,
- Enviar mensaje con **try-except** para manejo de error en valor ingresado,
- Esperar nuevamente los datos en caso de enviar mensaje de ERROR,
- Validar dato *unidad* con **upper** para convertir a may√∫scula,
- Enviar mensaje con **if-raise** para manejo de error y validar el ingreso de valor solicitado (C, F o K),
- Crear diccionario *conversiones* con f√≥rmulas de conversi√≥n de temperatura, donde unidad es la clave y valor el c√°lculo de la transformaci√≥n a otras unidades, 
- Almacenar resultado en *conversion_temp*: en dicc *conversiones* pedir la *unidad* (clave) en que tenemos el dato, se obtiene el valor que es las otras unidades de temperatura,
- Informar resultado de la conversi√≥n con **print** y devolverlo con **return**.


```python
def conversorTemperatura():
    print("Conversi√≥n de unidades de Temperatura")                                           
    try:
        temp = float(input("Introduce el valor numerico de la temperatura que deseas convertir:"))      # Solicitud datos                                                     # Ingreso temperatura  
    except ValueError:                                                                                  # Manejo de error si no es numerico
        print("ERROR, ingresa un valor num√©rico para la temperatura. Separador decimal con punto.")
        temp = float(input())                                                                           # Espera entrada nuevamente
    print("Indica en que unidades esta la temperatura (C para Celsius, F para Fahrenheit, K para Kelvin):")
    unidad = input().upper()                                                                            # Ingreso unidad de temp, convierte a mayusculas
    if unidad not in ["C", "F", "K"]:
        raise print("ERROR, ingresa una unidad v√°lida, debe ser C, F o K.")                             # Manejo de error si no es una unidad valida
    conversiones = { "C": {"F": round((temp * 9/5) + 32, 2), "K": round(temp + 273.15, 2)},                 
                     "F": {"C": round((temp - 32) * 5/9, 2), "K": round((temp - 32) * 5/9 + 273.15, 2)},# Diccionario con formulas para conversiones
                     "K": {"C": round(temp - 273.15, 2), "F": round((temp - 273.15) * 9/5 + 32, 2)}         
                }
    conversion_temp = conversiones[unidad]                                              # Busca en dicc unidad a convertir para calculo
    print(f"La temperatura",temp, unidad,"en otras unidades es:")
    return conversion_temp                                                              # Retorna las conversiones  

conversorTemperatura()
```

<font color=royalblue>3. Contador de palabras de un texto.</font>

Escribe una funci√≥n que dado un texto, muestre las veces que aparece cada palabra. Intenta que se gestionen todas las posibles casu√≠sticas que hagan que el programa no funcione correctamente.
(EXTRA): ¬øCu√°l es la longitud media de las palabras del texto que has escrito? ‚Äú¬øHola c√≥mo va?‚Äù deber√≠a devolver (4+3+2) / 3 = 3

<font color=royalblue>Objetivo - CONTADOR DE PALABRA:</font>
- Introducir objetivo de la funcion con un **print**,
- Ingresar de dato como variable *texto_string* con **input**,
- Convertir *texto_string* a lista en vriable *texto* con **split**,
- Normalizar variable *texto* con **lower**,
- Almacenar en variable *num_palabras* el n√∫mero de palabras en *texto* con **len**,
- Crear diccionario vacio en variable *palabras*, donde se contaran el numero de veces de cada palabra:
    - **for** para recorrer cada *palabra* en *lista*,
    - el condicional **if-else** para adicionar al diccionario cada *palabra* como clave y el n√∫mero de veces de una palabra como valor (+=1 √≥ =1 seg√∫n sea el caso),
- Enviar mensaje con resultados del diccionario *palabras* con **print**,
- Calcular longitud media de *palabra* en *lista*: con **for** para recorrer *lista* y determinar el acumulado de letras total.
    - Definir valor inicial en variable *acum_letras* = 0,  
    - Usar **range** para establecer la cantidad de iteraciones seg√∫n el n√∫mero de palabras en *texto* (*num_palabras*),
    - Almacenar en variable *num_letras* la longitud de cada *palabra* con **len**,
    - Finalizar el **for** adicionando en variable *acum_letras* el *num_letras* de cada *palabra*, 
- Calcular finalmente la logitud media dividiendo el total de letras entre num de palabras.

```python
def contarPalabras():
    print("Contador de palabras en texto:")   
    texto_string = input("Introduce el texto que quieres analizar:")                # Solicita dato 
    texto = texto_string.lower().split()                                            # Convierte a lista 
    num_palabras = len(texto)                                                       # Cuenta numero de palabras
    palabras = {}                                                                   # Diccionario vacio para conteo de palabras
    for palabra in texto:                                                           # Recorre lista de palabras
        if palabra in palabras:
            palabras[palabra] += 1                                                  # Si palabra existe, suma 1 al conteo    
        else:
            palabras[palabra] = 1                                                   # Si no existe, la agrega con conteo 1    
    print("El texto: '"+texto_string+"', tiene",num_palabras,"palabras,")
    print("distribuidas asi:",palabras)
    acum_letras = 0                                                                 # valor inicial de Acumulador de letras en lista
    for palabra in range (0, num_palabras):                                         # Recorre lista de palabras 
        num_letras= len(texto[palabra])                                             # Cuenta letras por palabra  
        acum_letras = acum_letras + num_letras                                      # Acumula total de letras  
    print("La longitud media de letras por palabra del texto es:",round(acum_letras/num_palabras,2),"letras")  # Calcula longitud media por palabra


contarPalabras()
```

Resulta que el cliente tiene una encuesta muy antigua que se almacena en un diccionario y los resultados los necesita a la inversa, es decir, intercambiando las claves y valores. Los valores y claves en el diccionario original son √∫nicos; si √©ste no es el caso, la funci√≥n deber√≠a imprimir un mensaje de advertencia, junto con una lista con los valores asociados a la clave repetida.

<font color=royalblue>Objetivo - DICC INVERSO:</font>  
- Introducir objetivo de la funci√≥n con un **print**,
- Crear diccionario en variable *dicc* al que se aplicar√° la conversi√≥n,
- Crear diccionario vacio en variable *dicc_inverso* para almacenar el intercambio a realizar,
- Llenar *dicc_inverso* con los datos de clave por valor y viceversa:
- Usar **for** para recorrer las variables clave y valor de *dicc* con el m√©todo **.item()**,
- Usar el condicional **if-else** para insertar en el diccionario *dicc_inverso* los datos al inverso, es decir valor: clave,
    - Con **if** crea el item en caso de no existir la clave =*valor* en el *dicc_inverso* (**not in**) con **update**, 
    - Con **else** maneja el caso en que ya exista la clave en *dicc:inverso*, de manera que debe crear una lista que adiccione valores para una misma clave, entonces dicc_inverso[valor] seran el valor o los valores dentro del diccionario inverso, donde se adicciona en lista el nuevo valor=*clave*. 
- Finalizar enviado mensaje con **print** y resultados del *dicc_inverso* con **return**. 
dicc.items()  

```python
def invertirDiccionario(): 
    print("Inversion de diccionario clave-valor")
    dicc = {'a': 1, 'b': 2, 'c': 1}                                 # Diccionario inicial
    dicc_inverso = {}                                               # Diccionario vacio para invertir
    for clave, valor in dicc.items():                               # Recorre el diccionario inicial para aplicar inversion clave-valor
        if valor not in dicc_inverso:                               # Si la clave no esta en el dicc inverso se agrega el par clave-valor
            dicc_inverso.update({valor: clave})
        else:                                                       # Si la clave ya existe, se agrupan las claves en lista
            dicc_inverso[valor] = [dicc_inverso[valor], clave]      
    print("El diccionario a la inversa quedaria asi:")
    return dicc_inverso                                             # Retorna diccionario invertido                        

invertirDiccionario()
```
